```python
#3. Perform clustering of Employees who left based on their satisfaction and evaluation.
#3.1.Choose columns satisfaction_level, last_evaluation and left.
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('1688640705_hr_comma_sep.csv');
data_left = data[data['left'] == 1]
data_left_cluster = data_left[['satisfaction_level', 'last_evaluation']]
scaler = StandardScaler();
data_left_cluster_scaled = scaler.fit_transform(data_left_cluster)
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(data_left_cluster_scaled)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method For Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#1.	Handle the left Class Imbalance using SMOTE technique.
#1.1.	Pre-Process the data by converting categorical columns to numerical columns by
#Separating categorical variables and numeric variables.
#Applying get_dummies() to the categorical variables.
#Combining categorical variables and numeric variables.
#Do the stratified split of the dataset to train and test in the ratio 80:20 with random_state=123.
#Upsample the train dataset using SMOTE technique from the imblearn module.

import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler

# Load the dataset
data = pd.read_csv('1688640705_hr_comma_sep.csv')

# Identify categorical and numerical columns
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
numerical_columns.remove('left')

# Apply get_dummies to categorical variables
data_categorical = pd.get_dummies(data[categorical_columns])

# Standardize numerical columns
scaler = StandardScaler()
data_numerical = pd.DataFrame(scaler.fit_transform(data[numerical_columns]), columns=numerical_columns)

# Combine categorical and numerical variables
data_processed = pd.concat([data_numerical, data_categorical, data['left']], axis=1)

# Separate features and target
X = data_processed.drop('left', axis=1)
y = data_processed['left']

# Stratified split of the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)

# Apply SMOTE to the training data
smote = SMOTE(random_state=123)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Print class distributions
print("Training set class distribution before SMOTE:\n", y_train.value_counts())
print("Test set class distribution:\n", y_test.value_counts())
print("Training set class distribution after SMOTE:\n", pd.Series(y_train_resampled).value_counts())
```

```python
#5.	Perform 5-Fold cross-validation model training and evaluate performance. 
#5.1.	Train a Logistic Regression model and apply a 5-Fold CV and plot the classification report.
#5.2.	Train a Random Forest Classifier model and apply the 5-Fold CV and plot the classification report.
#5.3.	Train a  Gradient Boosting Classifier model and apply the 5-Fold CV and plot the classification report.

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('1688640705_hr_comma_sep.csv')

# Identify categorical and numerical columns
categorical_columns = data.select_dtypes(include=['object']).columns.tolist()
numerical_columns = data.select_dtypes(include=['number']).columns.tolist()
numerical_columns.remove('left')

# Apply get_dummies to categorical variables
data_categorical = pd.get_dummies(data[categorical_columns])

# Standardize numerical columns
scaler = StandardScaler()
data_numerical = pd.DataFrame(scaler.fit_transform(data[numerical_columns]), columns=numerical_columns)

# Combine categorical and numerical variables
data_processed = pd.concat([data_numerical, data_categorical, data['left']], axis=1)

# Separate features and target
X = data_processed.drop('left', axis=1)
y = data_processed['left']

# Stratified split of the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)

# Apply SMOTE to the training data
smote = SMOTE(random_state=123)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Function to perform 5-Fold CV and plot classification report
def evaluate_model(model, X, y):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
    y_pred = cross_val_predict(model, X, y, cv=skf)
    report = classification_report(y, y_pred, output_dict=True)
    report_df = pd.DataFrame(report).transpose()

    print(classification_report(y, y_pred))

    plt.figure(figsize=(10, 6))
    sns.heatmap(report_df.iloc[:-1, :].T, annot=True, cmap='viridis')
    plt.title(f'Classification Report for {model.__class__.__name__}')
    plt.show()

# Logistic Regression
logistic_model = LogisticRegression(random_state=123)
evaluate_model(logistic_model, X_train_resampled, y_train_resampled)

# Random Forest Classifier
rf_model = RandomForestClassifier(random_state=123)
evaluate_model(rf_model, X_train_resampled, y_train_resampled)

# Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(random_state=123)
evaluate_model(gb_model, X_train_resampled, y_train_resampled)
```

```python
#6.	Identify the best model and justify the evaluation metrics used. 
#6.1.	Find the ROC/AUC for each model and plot the ROC curve.


from sklearn.metrics import roc_curve, auc, RocCurveDisplay

def plot_roc_auc(model, X, y):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
    tprs = []
    aucs = []
    mean_fpr = np.linspace(0, 1, 100)
    
    fig, ax = plt.subplots(figsize=(10, 6))

    for i, (train, test) in enumerate(skf.split(X, y)):
        model.fit(X[train], y[train])
        viz = RocCurveDisplay.from_estimator(
            model,
            X[test],
            y[test],
            name=f"Fold {i+1}",
            alpha=0.3,
            lw=1,
            ax=ax,
        )
        interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
        interp_tpr[0] = 0.0
        tprs.append(interp_tpr)
        aucs.append(viz.roc_auc)

    mean_tpr = np.mean(tprs, axis=0)
    mean_tpr[-1] = 1.0
    mean_auc = auc(mean_fpr, mean_tpr)
    std_auc = np.std(aucs)
    
    ax.plot(
        mean_fpr,
        mean_tpr,
        color="b",
        label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (mean_auc, std_auc),
        lw=2,
        alpha=0.8,
    )

    std_tpr = np.std(tprs, axis=0)
    tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
    tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
    ax.fill_between(
        mean_fpr,
        tprs_lower,
        tprs_upper,
        color="grey",
        alpha=0.2,
        label=r"$\pm$ 1 std. dev.",
    )

    ax.plot([0, 1], [0, 1], linestyle="--", lw=2, color="r", alpha=0.8)
    ax.set(
        xlim=[-0.05, 1.05],
        ylim=[-0.05, 1.05],
        title=f"ROC Curve for {model.__class__.__name__}",
    )
    ax.legend(loc="lower right")
    plt.show()

# Logistic Regression
plot_roc_auc(logistic_model, X_train_resampled.values, y_train_resampled.values)

# Random Forest Classifier
plot_roc_auc(rf_model, X_train_resampled.values, y_train_resampled.values)

# Gradient Boosting Classifier
plot_roc_auc(gb_model, X_train_resampled.values, y_train_resampled.values)
```

```python
#6.2.	Find the confusion matrix for each of the models.

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

def plot_confusion_matrix(model, X, y):
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)
    y_pred = cross_val_predict(model, X, y, cv=skf)
    cm = confusion_matrix(y, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='viridis')
    plt.title(f'Confusion Matrix for {model.__class__.__name__}')
    plt.show()

# Logistic Regression
plot_confusion_matrix(logistic_model, X_train_resampled.values, y_train_resampled.values)

# Random Forest Classifier
plot_confusion_matrix(rf_model, X_train_resampled.values, y_train_resampled.values)

# Gradient Boosting Classifier
plot_confusion_matrix(gb_model, X_train_resampled.values, y_train_resampled.values)
```

```python
#7.	Suggest various retention strategies for targeted employees.
#7.1.	Using the best model, predict the probability of employee turnover in the test data.

rf_model.fit(X_train_resampled, y_train_resampled)
y_pred_proba = rf_model.predict_proba(X_test)[:, 1]  # Probability of the positive class (turnover)

# Create a DataFrame to hold the actual values and predicted probabilities
results = pd.DataFrame({'Actual': y_test, 'Predicted_Probability': y_pred_proba})

# Display the first few rows of the results
print(results.head())
```

```python
#Based on the below probability score range, categorize the employees into four zones and suggest your thoughts on the retention strategies for each zone.
#Safe Zone (Green) (Score < 20%)
#Low Risk Zone (Yellow) (20% < Score < 60%)
#Medium Risk Zone (Orange) (60% < Score < 90%)
#High Risk Zone (Red) (Score > 90%).

def categorize_risk(prob):
    if prob < 0.2:
        return 'Safe Zone (Green)'
    elif 0.2 <= prob < 0.6:
        return 'Low Risk Zone (Yellow)'
    elif 0.6 <= prob < 0.9:
        return 'Medium Risk Zone (Orange)'
    else:
        return 'High Risk Zone (Red)'

results['Risk_Zone'] = results['Predicted_Probability'].apply(categorize_risk)

# Display the distribution of employees in each risk zone
print(results['Risk_Zone'].value_counts())

# Plot the distribution of employees in each risk zone
sns.countplot(data=results, x='Risk_Zone', palette=['green', 'yellow', 'orange', 'red'])
plt.title('Distribution of Employees in Each Risk Zone')
plt.xlabel('Risk Zone')
plt.ylabel('Number of Employees')
plt.show()
```
